# Regresión lineal

## 1. La forma básica de un modelo de regresión lineal simple es:

$$
y = \beta_0 + \beta_1 x + \epsilon
$$

Donde:
- \( y \) es la variable dependiente.
- \( x \) es la variable independiente.
- \( $\beta_0$ \) es el intercepto.
- \( $\beta_1$ \) es la pendiente.
- \( $\epsilon$ \) es el error.

## 2. Definimos la suma de los cuadrados de los errores (SSE):

$$
SSE = \sum (y_i - (\beta_0 + \beta_1 x_i))^2
$$

Donde \( $y_i$ \) son los valores observados y \( $x_i$ \) son los valores de la variable independiente.

Para encontrar los valores óptimos de \( $\beta_0$ \) y \( $\beta_1$ \), tomamos las derivadas parciales de \( $SSE$ \) respecto a \( $\beta_0$ \) y \( $\beta_1$ \), y las igualamos a cero (minimización):

### Derivada con respecto a \( $\beta_0$ \)

$$
\frac{\partial SSE}{\partial \beta_0} = -2 \sum (y_i - (\beta_0 + \beta_1 x_i)) = 0
$$

Resolviendo para \( $\beta_0$ \):

$$
\sum y_i = n \beta_0 + \beta_1 \sum x_i
$$

### Derivada con respecto a \( $\beta_1$ \)

$$
\frac{\partial SSE}{\partial \beta_1} = -2 \sum (y_i - (\beta_0 + \beta_1 x_i)) x_i = 0
$$

Resolviendo para \( $\beta_1$ \):

$$
\sum y_i x_i = n \beta_0 \bar{x} + \beta_1 \sum x_i^2
$$

## 3. Resolviendo las Ecuaciones:

Con las dos ecuaciones obtenidas:

$$
\begin{align*}
\sum y_i &= n \beta_0 + \beta_1 \sum x_i \\
\sum y_i x_i &= n \beta_0 \bar{x} + \beta_1 \sum x_i^2
\end{align*}
$$

Podemos reescribir la primera ecuación para despejar \( $\beta_0$ \):

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

Donde \( $\bar{x}$ \) y \( $\bar{y}$ \) son las medias de \( $x$ \) y \( $y$ \), respectivamente.

Sustituyendo \( $\beta_0$ \) en la segunda ecuación y resolviendo para \( $\beta_1$ \):

$$
\beta_1 = \frac{n(\sum y_i x_i) - (\sum x_i)(\sum y_i)}{n(\sum x_i^2) - (\sum x_i)^2}
$$

## 4. Cálculo de \( $\beta_0$ \)

Una vez que tenemos \( \beta_1 \), podemos calcular \( \beta_0 \) utilizando la relación obtenida anteriormente:

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

## 5. Resumen 

Finalmente, los coeficientes son:

$$
\beta_1 = \frac{n(\sum y_i x_i) - (\sum x_i)(\sum y_i)}{n(\sum x_i^2) - (\sum x_i)^2}
$$

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

## 6. Interpretación

- **\( $\beta_1$ \)**: Representa el cambio promedio en \( $y$ \) por cada unidad de cambio en \( $x$ \).
- **\( $\beta_0$ \)**: Es el valor de \( $y$ \) cuando \( $x$ \) es cero, representando el intercepto de la línea en el eje \( $y$ \).


# Problema 1.

Dado el siguiente conjunto de datos:

| \(x\) | \(y\) |
|-------|-------|
| 1     | 2     |
| 2     | 4     |
| 3     | 5     |
| 4     | 4     |
| 5     | 5     |

queremos encontrar la ecuación de la recta de regresión lineal en la forma:

$$
y = \beta_0 + \beta_1 x
$$

donde \( $\beta_0$ \) es el intercepto y \( $\beta_1$ \) es la pendiente de la recta. Además, minimizaremos el error cuadrático medio (MSE) para este conjunto de datos.


# Solución problema 1

## Paso 1: Calcular las Sumas Necesarias

Calculamos las sumas necesarias para encontrar los coeficientes \( $\beta_0$ \) y \( $\beta_1$ \):

$$
\sum x_i, \quad \sum y_i, \quad \sum x_i^2, \quad \sum x_i y_i
$$

| \($x_i$\) | \($y_i$\) | \($x_i^2$\) | \($x_i y_i$\) |
|---------|---------|-----------|--------------|
| 1       | 2       | 1         | 2            |
| 2       | 4       | 4         | 8            |
| 3       | 5       | 9         | 15           |
| 4       | 4       | 16        | 16           |
| 5       | 5       | 25        | 25           |

Ahora sumamos cada columna:

$$
\sum x_i = 1 + 2 + 3 + 4 + 5 = 15
$$

$$
\sum y_i = 2 + 4 + 5 + 4 + 5 = 20
$$

$$
\sum x_i^2 = 1 + 4 + 9 + 16 + 25 = 55
$$

$$
\sum x_i y_i = 2 + 8 + 15 + 16 + 25 = 66
$$

## Paso 2: Calcular la Pendiente \( $\beta_1$ \)

$$
\beta_1 = \frac{n \sum x_i y_i - \sum x_i \sum y_i}{n \sum x_i^2 - (\sum x_i)^2}
$$

donde \( $n$ \) es el número de puntos (\( n = 5 \)).

Sustituyendo los valores:

$$
\beta_1 = \frac{5(66) - (15)(20)}{5(55) - (15)^2}
$$

$$
\beta_1 = \frac{330 - 300}{275 - 225}
$$

$$
\beta_1 = \frac{30}{50} = 0.6
$$

## Paso 3: Calcular el Intercepto \( $\beta_0$ \)

La fórmula para calcular el intercepto es:

$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$

donde:

$$
\bar{x} = \frac{\sum x_i}{n} = \frac{15}{5} = 3
$$

$$
\bar{y} = \frac{\sum y_i}{n} = \frac{20}{5} = 4
$$

Sustituyendo los valores de \( $\bar{x}$ \), \( $\bar{y}$ \) y \( $\beta_1$ \):

$$
\beta_0 = 4 - 0.6(3)
$$

$$
\beta_0 = 4 - 1.8 = 2.2
$$

## Paso 4: Ecuación de la Recta de Regresión

La ecuación de la recta de regresión lineal es:


$$
y = \beta_0 + \beta_1 x = 2.2 + 0.6x
$$


## Paso 5: Calcular el Error Cuadrático Medio (MSE)

El error cuadrático medio se calcula como:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

donde \( $\hat{y}_i = 2.2 + 0.6 x_i$ \) son los valores predichos.

Calculamos los valores de \( $\hat{y}_i$ \) y \( ($y_i - \hat{y}_i)^2$ \):

| \($x_i$\) | \($y_i$\) | \( $\hat{y}_i = 2.2 + 0.6 x_i$ \) | \( $(y_i - \hat{y}_i)^2$ \) |
|---------|---------|-------------------------------|---------------------------|
| 1       | 2       | 2.8                           | 0.64                      |
| 2       | 4       | 3.4                           | 0.36                      |
| 3       | 5       | 4.0                           | 1.00                      |
| 4       | 4       | 4.6                           | 0.36                      |
| 5       | 5       | 5.2                           | 0.04                      |

Ahora sumamos los cuadrados de los errores:

$$
\sum (y_i - \hat{y}_i)^2 = 0.64 + 0.36 + 1.00 + 0.36 + 0.04 = 2.40
$$

Finalmente, calculamos el MSE:

$$
\text{MSE} = \frac{2.40}{5} = 0.48
$$

## Conclusión

La ecuación de la recta de regresión lineal que mejor se ajusta a nuestros datos es:

$$
y = 2.2 + 0.6 x
$$

Y el error cuadrático medio (MSE) es \( 0.48 \).



# Código 1.

```python
import numpy as np
import matplotlib.pyplot as plt

# Conjunto de datos de entrenamiento
train_data = np.array([
    [1, 2],
    [2, 4],
    [3, 5],
    [4, 4],
    [5, 5]
])

# Separar los datos de entrenamiento en x e y
x_train = train_data[:, 0]
y_train = train_data[:, 1]

# 1. Obtención de los coeficientes beta_0 y beta_1
n_train = len(y_train)
x_mean_train = np.mean(x_train)
y_mean_train = np.mean(y_train)

# Calcular beta_1
beta_1 = np.sum((x_train - x_mean_train) * (y_train - y_mean_train)) / np.sum((x_train - x_mean_train) ** 2)

# Calcular beta_0
beta_0 = y_mean_train - beta_1 * x_mean_train

# Ecuación de la recta
print(f"Ecuación de la recta: y = {beta_0:.2f} + {beta_1:.2f}x")

# 2. Calcular el Error Cuadrático Medio (MSE) para el conjunto de entrenamiento
y_train_pred = beta_0 + beta_1 * x_train
mse_train = np.mean((y_train - y_train_pred) ** 2)
print(f"Error Cuadrático Medio (MSE) en entrenamiento: {mse_train:.2f}")

# 3. Gráfico de los Datos de Entrenamiento y la Recta de Regresión
plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, color='blue', label='Datos de Entrenamiento')
plt.plot(x_train, y_train_pred, color='red', label='Recta de Regresión (Entrenamiento)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión Lineal - Datos de Entrenamiento')
plt.legend()
plt.grid()
```

<div class="alert alert-block alert-info">
<b>

## Conclusión 1:

La gráfica muestra claramente la relación entre los puntos de datos y la recta de regresión que mejor se ajusta a los mismos. La recta de ajuste minimiza el error cuadrático medio (MSE), lo que garantiza una buena aproximación de la tendencia de los datos.

# Coeficiente de Determinación \( $R^2$ \)

El **Coeficiente de Determinación**, comúnmente denotado como \( $R^2$ \), es una medida estadística que representa la proporción de la variabilidad total de una variable dependiente que puede ser explicada por una o más variables independientes en un modelo de regresión.

## Definición Formal

El \( $R^2$ \) se calcula como:

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}} 
$$

Donde:
- \( $SS_{res}$ \) es la suma de los cuadrados de los residuos, que mide la variabilidad que no es explicada por el modelo.
- \( $SS_{tot}$ \) es la suma total de los cuadrados, que mide la variabilidad total de la variable dependiente.

## Interpretación

- **Valor de \( $R^2$ \)**:
  - \( $R^2$ = 0 \): Significa que el modelo no explica ninguna de las variaciones de la variable dependiente.
  - \( $R^2$ = 1 \): Indica que el modelo explica todas las variaciones de la variable dependiente.
  - \( $0 < R^2 < 1$ \): Refleja la proporción de variabilidad explicada por el modelo. Por ejemplo, un \( $R^2 = 0.80$ \) sugiere que el 80% de la variabilidad de la variable dependiente se explica por el modelo.

## Consideraciones

- **No siempre un alto \( $R^2$ \) significa un buen modelo**: Un \( $R^2$ \) alto no garantiza que el modelo sea adecuado o que las relaciones sean significativas. Es importante evaluar otros indicadores y realizar un análisis exhaustivo del modelo.
- **Ajuste del modelo**: Al agregar más variables independientes a un modelo, \( $R^2$ \) generalmente no disminuye, lo que puede llevar a sobreajuste. Por esta razón, se utiliza el **\( $R^2$ \) ajustado**, que penaliza la inclusión de variables innecesarias.

Para calcular el coeficiente de determinación \( R^2 \) en un problema de regresión lineal, necesitas realizar los siguientes pasos:

## 1. Calcular la media de la variable dependiente \( $y$ \)

$$ 
\bar{y} = \frac{\sum y_i}{n} 
$$

donde \( $n$ \) es el número de observaciones.

## 2. Calcular la Suma Total de Cuadrados (SST)

$$
SST = \sum (y_i - \bar{y})^2 
$$

## 3. Calcular la regresión para obtener \( $y$ \) predicho ($\hat{y}$) y luego calcular la Suma de Cuadrados de los Errores (SSE)


$$
SSE = \sum (y_i - \hat{y}_i)^2 
$$

## 4. Calcular \( $R^2$ \)

$$
R^2 = 1 - \frac{SSE}{SST} 
$$

# Solución problema 2.

# Cálculo del Coeficiente de Determinación \( R^2 \)

$$

## 2. Calcular la Suma Total de Cuadrados (SST)

La suma total de cuadrados se calcula como:

$$
SST = \sum (y_i - \bar{y})^2 = (2 - 4)^2 + (4 - 4)^2 + (5 - 4)^2 + (4 - 4)^2 + (5 - 4)^2
$$

$$
SST = (-2)^2 + (0)^2 + (1)^2 + (0)^2 + (1)^2 = 4 + 0 + 1 + 0 + 1 = 6
$$

$$

5. Calcular el Coeficiente de Determinación $R^2$
	

$$
R^2 = 1 - \frac{SSE}{SST} = 1 - \frac{2.4}{6} = 1 - 0.4 = 0.6

### Conclusión
	
Esto indica que el $60\%$ de la variabilidad en \( $y$ \) se puede explicar por la variabilidad en \( $x$ \) a través del modelo de regresión lineal.

```python
# Cálculo del Coeficiente de Determinación R^2 para el conjunto de entrenamiento
sst_train = np.sum((y_train - y_mean_train) ** 2)  # Suma total de cuadrados
sse_train = np.sum((y_train - y_train_pred) ** 2)   # Suma de cuadrados de errores
r_squared_train = 1 - (sse_train / sst_train)
print(f"Coeficiente de Determinación R^2 en entrenamiento: {r_squared_train:.2f}")

# Mostrar el gráfico
# 3. Gráfico de los Datos de Entrenamiento y la Recta de Regresión
plt.figure(figsize=(10, 6))
plt.scatter(x_train, y_train, color='blue', label='Datos de Entrenamiento')
plt.plot(x_train, y_train_pred, color='red', label='Recta de Regresión (Entrenamiento)')
plt.text(0.5, 0.5, f'$R^2 = {r_squared_train:.2f}$', fontsize=15, color='green', transform=plt.gca().transAxes)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión Lineal - Datos de Entrenamiento')
plt.legend()
plt.grid()
plt.show()
```

# Problema de clasificación (Regresión logística)

La curva ROC (Receiver Operating Characteristic) es una representación gráfica que ilustra el rendimiento de un clasificador a través de diferentes umbrales de decisión. 

Se utiliza principalmente en problemas de clasificación binaria.

1. Calcular Predicciones
	
Calcular las predicciones $\hat{y}$ para los puntos del conjunto de prueba utilizando la ecuación de la recta ajustada. <span style="color:green;">&#10004;</span> 

2. Definir un Umbral
	
Definir un umbral para clasificar las predicciones como positivas o negativas. Por ejemplo, si $\hat{y}$ es mayor que un valor específico, se clasifica como positivo.

3. Crear la Matriz de Confusión
	
Comparar las predicciones con las etiquetas verdaderas para construir la matriz de confusión.


|            | Predicción 0 | Predicción 1 |
|------------|---------------|--------------|
| **Real 0** | TN            | FP           |
| **Real 1** | FN            | TP           |

- **TN (True Negative)**: Número de verdaderos negativos.
- **FP (False Positive)**: Número de falsos positivos.
- **FN (False Negative)**: Número de falsos negativos.
- **TP (True Positive)**: Número de verdaderos positivos.

# Calcular las Métricas

## 1. Cálculo de la Curva ROC

Para calcular la curva ROC, necesitamos determinar la Tasa de Verdaderos Positivos (TPR) y la Tasa de Falsos Positivos (FPR) para varios umbrales.

- **TPR (Sensibilidad)**: Proporción de verdaderos positivos identificados correctamente.

$$
TPR = \frac{TP}{TP + FN}
$$

- **FPR**: Proporción de falsos positivos identificados incorrectamente.

$$
FPR = \frac{FP}{FP + TN}
$$


## 2. Para calcular el F1-score, utilizamos la siguiente fórmula:

$$
F_1 = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
$$

Donde:

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

y

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

Aquí:
- \( TP \) es el número de verdaderos positivos.
- \( FP \) es el número de falsos positivos.
- \( FN \) es el número de falsos negativos.

Recall

Recall = \frac{TP}{TP + FN} = \frac{1}{1 + 0} = 1

F1-score

\begin{align*}
F1 &= 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall} \\

```python

# Valores reales y predicciones
x = np.array([1, 2.5, 3.8, 4, 6])
y_real = np.array([2, 3, 5, 4, 6])

# Etiquetas verdaderas 
verdaderos = np.array([0, 0, 0, 0, 1])
y_pred = 2.2 + 0.6 * x

# Etiquetas predichas
predichos = []
for i in range(len(y_pred)):
    if y_pred[i] > 4:
        predichos.append(1)
    else:
        predichos.append(0)
        
# Matriz de confusión
TP = 0
FP = 0
FN = 0
TN = 0

for i in range(len(verdaderos)):
    if verdaderos[i] == 1 and predichos[i] == 1:
        TP += 1
    elif verdaderos[i] == 0 and predichos[i] == 1:
        FP += 1
    elif verdaderos[i] == 1 and predichos[i] == 0:
        FN += 1
    else:
        TN += 1

# 
print(f"Matriz de Confusión:")
print(f"TP: {TP}, FP: {FP}, FN: {FN}, TN: {TN}")

# Precisión

precision = TP / (TP + FP)
print(f"Precisión: {precision:.2f}")

# Recall
recall = TP / (TP + FN)
print(f"Recall: {recall:.2f}")

# F1-Score
f1_score = 2 * (precision * recall) / (precision + recall)
print(f"F1-Score: {f1_score:.2f}")

# Curva ROC
# Calcular la tasa de verdaderos positivos (TPR) y la tasa de falsos positivos (FPR)

TPR = TP / (TP + FN)
print(f"TPR: {TPR:.2f}")
FPR = FP / (FP + TN)
print(f"FPR: {FPR:.2f}")

# Graficar la curva ROC
plt.figure(figsize=(6, 6))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(FPR, TPR, 'ro')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC')

plt.show()

```
