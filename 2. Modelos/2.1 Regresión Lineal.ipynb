Es un método estadístico que se utiliza para modelar la relación entre dos variables a través de una línea recta. Esta línea recta se utiliza para predecir el valor de una variable desconocida, en función del valor de una variable conocida.

Por ejemplo:

                  Precio de la casa = b + m * tamaño de la casa

Existen 2 tipos de regresión lineal:

## Simple:

Implica solo dos variables, una dependiente y una independiente. La variable independiente se utiliza para predecir la variable dependiente, mediante una línea recta.

                y = β₀+β₁*x + ε

donde:

**y** es la variable independiente.

**β** son dos constantes desconocidas que representan:
1. $\beta_0$: el punto de intersección 
2. $\beta_1$: la pendiente

**ε**  es la función de pérdida.

## Múltiple

Involucra mas variables independientes para predecir una variable dependiente. En este mopdelo, se ajusta una ecuación de regresión que explica la relación entre la variable dependiente y múltiples variables independientes.

            y = β₀ + β₁x₁ + β₂x₂ +…+ βₐxₐ + ε 
            Y = β X (producto matricial)

donde: 

**y** es la variable dependiente.

**x** es una variable independiente.

**β** son coeficientes.

**ε** es la función de pérdida.



## Supuesto de linealidad

**Linealidad**: La relación de las variables debe poder representarse con una línea recta.

**Indepenencia**: El erro en una observación no debe estar relacionado con el error en otra observación.

**Homocedasticidad**: Los errores deben tener varianzas iguales, es decir, la dispersión de los errores debe ser constante.

**Normalidad**: Los errores deben seguir una distribución normal, es decir, la mayoría de los errores deben estar cerca de cero, así mismo su disprersión

## Evaluación de la regresión lineal

Las técnicas comúnes para evaluar la regresión lineal, incluyen el error cuadrático medio, el coeficiente de determinación (R²), el análisis de residuos, la prueba de significancia y la validación cruzada. Estas técnicas pueden ayudar a determinar si el modelo de regresión lineal se ajusta bien a los datos y si se adecuado para su uso.

1. Coeficiente de Determinación  $R^2$

Mide la proporción de la variabilidad en la variable dependiente que se puede explicar mediante la variabilidad en las variables independientes. Se calcula como:

$
R^2 = 1 - \frac{\text{Suma de los errores cuadráticos residuales (SSE)}}{\text{Suma total de los cuadrados (SST)}},
$

* SSE (Sum of Squared Errors) es la suma de los cuadrados de los residuos, es decir, la diferencia entre los valores observados y los valores predichos por el modelo.
* SST (Total Sum of Squares) es la suma de los cuadrados de la diferencia entre los valores observados y el valor promedio de los datos.

El valor de $R^2$ varía entre 0 y 1, donde:

* $R^2$ = 1  indica que el modelo explica perfectamente la variabilidad en los datos.
* $R^2$ = 0 indica que el modelo no explica la variabilidad en los datos.

En el caso particular de una regresión lineal, $R^2$ coincide con el cuadrado del coeficiente de correlación de Pearson.

2. Error Cuadrático Medio (MSE) y Raíz del Error Cuadrático Medio (RMSE)
Estas métricas miden la magnitud del error promedio en las predicciones del modelo:

Error cuadrático medio (MSE) se define como:

$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$

donde $y_i$ es el valor observado, $\hat{y}_i$ es el valor predicho por el modelo y $n$ es el número de observaciones.
    
Raíz del error cuadrático medio (RMSE) se obtiene tomando la raíz cuadrada del MSE:

$RMSE=\sqrt(MSE)$

Estas métricas dan una idea de la magnitud promedio de los errores en las predicciones. Un RMSE más bajo indica un mejor ajuste del modelo.


3. Error Absoluto Medio (MAE)
El error absoluto medio mide la media de los errores absolutos entre los valores observados y los predichos. Se calcula como:

$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|.
$

4. Visualización de los Residuos

Los residuos representan el error de predicción del modelo para cada dato específico. Si un residuo es positivo significa que el valor observado es mayor que el valor predicho por el modelo. Si es negativo significa que el valor observado es menor que el valor predicho.

En la regresión lineal, se espera que los residuos estén distribuidos aleatoriamente alrededor de cero. Esto indicaría que el modelo captura correctamente la relación entre las variables independientes y la variable dependiente.

Los residuos son fundamentales para evaluar la calidad del ajuste del modelo de regresión. Algunas técnicas para analizar los residuos incluyen:

El residuo para una observación $i$ se calcula como:

$
e_i = y_i - \hat{y_i}
$

* $y_i$ es el valor observado (real) de la variable dependiente para la observación $i$
* $\hat{y_i}$ es el valor predicho por el modelo de regresión para la observación $i$
* $e_i$ es el residuo correspondiente a la observación $i$

Normalidad de los Residuos

En la regresión lineal, se asume que los residuos siguen una distribución normal. Una forma de verificar esta suposición es con un gráfico Q-Q (quantile-quantile) o mediante pruebas de normalidad, como la prueba de Shapiro-Wilk.
Heterocedasticidad

La heterocedasticidad ocurre cuando la varianza de los residuos no es constante a lo largo de los valores predichos. Esto puede indicar problemas en el modelo de regresión. Se puede detectar al observar un patrón en el gráfico de residuos vs. valores predichos.

El Error Cuadrático Medio (MSE) se calcula usando los residuos. Es la media de los cuadrados de los residuos y mide la magnitud promedio del error de predicción:

 $
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} e_i
$

5. Evaluación con Validación Cruzada

La validación cruzada es una técnica que se utiliza para evaluar el modelo dividiendo los datos en subconjuntos, por ejemplo, en $k-partes$ o $k-fold$. El modelo se entrena con algunos subconjuntos y se evalúa con los otros. Esto proporciona una mejor estimación del rendimiento del modelo en datos no vistos.


# Ejemplo de Regresión Lineal Simple

## Scikit learn: Linear Regression

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Generamos datos sintéticos
def f(x):  # función f(x) = 2*x + 10 + 20*ruido
    np.random.seed(42)
    y = 2*x + 10 + 20*np.random.randn(x.shape[0])
    return y
x = np.arange(0, 200, 5) 
y = f(x) 
plt.scatter(x,y,label='data', color='blue')
plt.title('Datos')
plt.xlabel('x')
plt.ylabel('f(x) = 2*x + 10 + 0.2*ruido')
```

```python
# Importamos la clase de Regresión Lineal de scikit-learn
from sklearn.linear_model import LinearRegression 
# Importamos las funciones de error cuadrático medio, raíz del error cuadrático medio y coeficiente de determinación
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Creamos un objeto de la clase LinearRegression
regresion_lineal = LinearRegression()
# Entrenamos el modelo de regresión lineal
# reshape(-1,1) es para convertir un array de numpy en un vector columna
regresion_lineal.fit(x.reshape(-1,1), y) 
# Obtenemos la predicción
prediccion_entrenamiento = regresion_lineal.predict(x.reshape(-1,1)) 

# Error Cuadrado Medio
mse = mean_squared_error(y_true = y, y_pred = prediccion_entrenamiento)
# La raíz cuadrada del MSE es el RMSE
rmse = np.sqrt(mse)
# calculamos el coeficiente de determinación R2
r2 = regresion_lineal.score(x.reshape(-1,1), y)

print('m = ' + str(regresion_lineal.coef_) + ', b = ' + str(regresion_lineal.intercept_)) # parámetros que ha estimado la regresión lineal
print('Error Cuadrático Medio (MSE) = ' + str(mse) + ' , Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))
print('Coeficiente de Determinación R2 = ' + str(r2))
plt.scatter(x, y, color="red")
plt.plot(x, prediccion_entrenamiento, color="blue")
plt.title('Datos')
plt.xlabel('x')
plt.ylabel('y')
```

### Métricas de Evaluación

Un RMSE más bajo indica un mejor ajuste del modelo.

El R² es una medida de la proporción de la varianza en la variable dependiente que es explicada por la variable independiente. Un R² de 0.9736 indica que el 97.36% de la variabilidad en la variable dependiente es explicada por el modelo.
Un R² cercano a 1 indica un ajuste muy bueno del modelo a los datos.

### Otros parámetros del modelo de regresión lineal de sci-kitlearn:


* **fit\_intercept**: Si es *True*, el modelo calculará el intercepto $\beta_0$. Si es *False*, no se incluirá el intercepto en el modelo.
* **normalize**: Si es *True*, las características se normalizarán antes de ajustar el modelo. Este parámetro es relevante si **fit_intercept** es *True*.
* **copy_X**: Si es *True*, se hace una copia de los datos. Si es *False*, los datos originales pueden ser modificados.

# Regresión Lineal con base de datos (una característica)

# Importar datos

https://www.kaggle.com/code/pd1994/salary-data-simple-linear-regression

```python
import pandas as pd
import os

mainpath = "./datasets/"
filename = 'salary_data.csv'
fullpath = os.path.join(mainpath, filename)
dataset= pd.read_csv (fullpath)
dataset.head()
```

# Seleccionar características

```python
target = dataset['Salary'] # variable dependiente
features = dataset['YearsExperience'] # variable independiente

target = target.values.reshape(-1,1)
features = features.values.reshape(-1,1)
```

# Separar datos en datos de entrenamiento y datos de prueba

Los **datos de entrenamiento** son el subconjunto más grande del conjunto de datos original y se utiliza para entrenar o ajustar el modelo de aprendizaje automático. Los datos de entrenamiento varían dependiendo de si utilizamos algoritmos de aprendizaje supervisado o de aprendizaje no supervisado.

Para el aprendizaje no supervisado, los datos de entrenamiento contienen puntos de datos sin etiquetar, es decir, las entradas no están etiquetadas con las salidas correspondientes. Se requieren modelos para encontrar los patrones de los conjuntos de datos de entrenamiento dados para poder hacer predicciones.

Para el aprendizaje supervisado, los datos de entrenamiento contienen etiquetas para entrenar el modelo y realizar predicciones.

Los **datos de prueba** evalúan el rendimiento del modelo y garantiza que el modelo pueda generalizarse bien con el conjunto de datos nuevo. Este conjunto de datos es independiente del conjunto de datos de entrenamiento, pero tiene algunos características y distribución de probabilidad similares a las de los datos de entrenamiento y los utiliza como punto de referencia para la evaluación del modelo una vez que se completa el entrenamiento del modelo. Por lo general, el conjunto de datos de prueba representa aproximadamente entre el 20 y el 25 % del total de datos originales.

```python
#División de datos para entrenamiento y prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)  
```

# Entrenar el modelo

```python
#Crear modelo de regresión lineal simple con el conjunto de entrenamiento
model = LinearRegression()
#Ajustar el modelo usando el modelo de la clase (debe tener mismo numero de filas tanto x como y)
model.fit(X_train, y_train)
#Creando un vector de predicciones, se debe tomar solo los valores independientes
y_pred = model.predict(X_test)

# Error Cuadrado Medio
mse = mean_squared_error(y_test, y_pred)
# La raíz cuadrada del MSE es el RMSE 
rmse = np.sqrt(mse)
# Coeficiente de determinación R2
r2 = r2_score(y_test, y_pred)
# Error absoluto medio
mae = mean_absolute_error(y_test, y_pred)**0.5


print('m = ' + str(model.coef_) + ', b = ' + str(model.intercept_)) # parámetros que ha estimado la regresión lineal
print('Error Cuadrático Medio (MSE) = ' + str(mse))
print('Raíz del Error Cuadrático Medio (RMSE) = ' + str(rmse))
print('Coeficiente de Determinación R2 = ' + str(r2))
print('Error Absoluto Medio (MAE) = ' + str(mae))

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted Values')
plt.text(0.05, 0.95, f'RMSE: {rmse:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.text(0.05, 0.90, f'Coefficients: {model.coef_}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.show()
```

```python
def linear_regression_model(data,target_column):
    '''
    Esta función entrena un modelo de regresión lineal y calcula el error cuadrático medio.
    
    Parámetros:
        data: DataFrame con las características y la variable objetivo
    
    1. Separar las características (f0, f1, f2,) y la variable objetivo (product)
        target = data['y']
        features = data.drop(['y'], axis=1)
    2. Dividir los datos en conjuntos de entrenamiento (75%) y prueba (25%)
        X: características
        Y: variable objetivo
        X_train, y_train: Conjunto de entrenamiento
        X_test, y_test: Conjunto de prueba
    3. Entrenar el modelo de regresión lineal
        model = LinearRegression()
        model.fit(X_train, y_train)
    4. Obtener la predicción del modelo de regresión lineal ya entrenado
        y_pred = model.predict(X_test)
    5. Calcular el error cuadrático medio
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)**0.5
    6. Retornar la predicción del modelo entrenado y el error cuadrático medio
        return model, mse, rmse, r2, mae, y_test, y_pred
    '''
   
    features = data.drop(columns=[target_column])
    target = data[target_column]

    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)**0.5

    print(f'Model Coefficients: {model.coef_}')
    print(f'Model Intercept: {model.intercept_}')
    print('Mean Squared Error: ', mse)
    print('Root Mean Squared Error: ', rmse)
    print('Mean Absolute Error: ', mae)
    print('R2 Score: ', r2)


    plt.figure(figsize=(10, 6))
    plt.scatter(y_test, y_pred, alpha=0.7)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
    plt.xlabel('Actual')
    plt.ylabel('Predicted')
    plt.title('Actual vs Predicted Values')
    plt.text(0.05, 0.95, f'RMSE: {rmse:.2f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
    plt.text(0.05, 0.90, f'Coefficients: {model.coef_}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
    plt.show()
    #return model, mse, rmse, r2, mae, y_test, y_pred
```
