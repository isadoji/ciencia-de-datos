# Exploraci√≥n de las atransformaciones de los datos a sqrt, log y box-cox

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd

from scipy.stats import boxcox 
#https://www.kaggle.com/datasets/rakannimer/air-passengers?select=AirPassengers.csv

df = pd.read_csv('../datasets/AirPassengers.csv',  index_col = 'Month', parse_dates=True)
df.head()
```

```python
df['#Passengers'].plot(figsize=(12, 8))
```

## Caracter√≠sticas

1. Creciente
2. Sinusoidal creciente (amplitud)

```python
df['Sqrt_Passengers'] = np.sqrt(df['#Passengers'])
df['Sqrt_Passengers'].plot(figsize=(12, 8))
```

```python
df['Log_Passengers'] = np.log(df['#Passengers'])
df['Log_Passengers'].plot(figsize=(12, 8))
```

```python
df['Boxcox_Passengers'], lam = boxcox(df['#Passengers']) # lam is the lambda value
df['Boxcox_Passengers'].plot(figsize=(12, 8))
```

## Box-Cox

Transformaci√≥n de una distribuci{on no-normal a una normal, ajustando $\lambda$

```python
np.random.seed(0)
data = np.random.exponential(size=1000)
transformed_data, best_lambda = boxcox(data) 
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.hist(data, density=True,bins = 100, histtype = 'step')
plt.subplot(1,2,2)
plt.hist(transformed_data, density=True,bins = 100, histtype = 'step')
plt.show()
```
